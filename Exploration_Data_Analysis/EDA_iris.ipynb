```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python

# Load the Iris dataset
iris = load_iris()
data = iris.data
target = iris.target
target_names = iris.target_names
feature_names = iris.feature_names
```

```python

# Create a DataFrame for the Iris data
iris_df = pd.DataFrame(data, columns=feature_names)
iris_df['species'] = target_names[target]
```

```python
# Explore the data
print(iris_df.head())
print(iris_df.info())
print(iris_df.describe())
```

```python

# Visualize the data
# Pairplot to see the relationships between features
sns.pairplot(iris_df, hue='species', markers=['o', 's', 'D'])
plt.title('Pairplot of Iris Dataset')
plt.show()
```

```python
# Box plots to see the distribution of features for each species
plt.figure(figsize=(10, 6))
for i, feature in enumerate(feature_names):
    plt.subplot(2, 2, i + 1)
    sns.boxplot(x='species', y=feature, data=iris_df)
    plt.title(f'Distribution of {feature} by Species')
plt.tight_layout()
plt.show()
```

```python
# Correlation heatmap to see the feature correlations
plt.figure(figsize=(8, 6))
sns.heatmap(iris_df.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
