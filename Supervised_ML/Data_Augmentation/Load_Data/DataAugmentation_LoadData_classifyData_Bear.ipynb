```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

Get the content

```python
data_path = '/content/Bear'
```

```python
image_size = (150, 150)
batch_size = 10
```

```python
train_datagen =  ImageDataGenerator(rescale=1./255, validation_split=0.2)

train_batch = train_datagen.flow_from_directory(
    data_path,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='binary',
    subset='training')

validation_batch = train_datagen.flow_from_directory(
    data_path,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='binary',
    subset='validation')

test_batch = train_datagen.flow_from_directory(
    data_path,
    target_size=image_size,
    batch_size=batch_size,
    class_mode='binary',
    subset='validation')
```

```python
imgs, labels = train_batch[0]
plt.figure(figsize=(18, 2))
for i in range(10):
  plt.subplot(1,10,1 + i)
  plt.axis('off')
  plt.imshow(imgs[i], cmap='gray_r')
plt.show()
print(labels[:10])
```
